### Convirtiendo a DateTimeüìÖ‚åõ

El m√≥dulo datetime incluye funciones y clases para hacer an√°lisis, formateo y aritm√©tica de fecha y hora

<strong> `datetime` </strong> contiene funciones y clases para trabajar con fechas y horas, por separado y en conjunto.


```python
#conviertiendo la columna endTime a formato de datetime
spotify_df['endTime'] = pd.to_datetime(spotify_df['endTime'])

spotify_df["endTime"].head()
```

```python
#Para organizar o listar por un tipo de frecuencia las fechas
spotify_df['Date'] = spotify_df['endTime'].dt.to_period('D')
spotify_df.head()
```

```python
#Otra forma de extraer el a√±o de la fecha
spotify_df['year'] = pd.DatetimeIndex(spotify_df['endTime']).year
spotify_df.head()
```

```python
#extraer el mes de la fecha
spotify_df['month'] = pd.DatetimeIndex(spotify_df['endTime']).month
spotify_df.head()
```

```python
#agg permite realizar agregaciones
spotify_df.groupby('year').agg({'msPlayed':['min','max','mean', 'sum', 'count']})
```

```python
# total de milisegundos escuhados en 2019 contra el total de los escuchados en 2020 para todo el dataset.
spotify_df.groupby('year').mean()["msPlayed"].plot(kind='bar', color=['royalblue','green'])
```

```python
#Contando el n√∫mero de canciones de cada uno de los artistas
ms_artist=(spotify_df.groupby(['artistName'])['msPlayed']).sum()
ms_artist.sort_values(ascending=False)[0:10]
```

```python
#Top 10 de artistas mas escuhados

plt.figure(figsize=(8,4))
ms_artist.sort_values(ascending=False)[0:10].sort_values().plot.barh(cmap='jet')

plt.xticks(rotation=50)
plt.title("Top 10 most listened to artists")
plt.xlabel("Milliseconds heard")
plt.ylabel("Artist Name")
plt.show()
```

```python
ms_track=(spotify_df.groupby(['trackName'])['msPlayed']).sum()
ms_track.sort_values(ascending=False)[0:10]
```

```python
#Top 10 de canciones mas escuchadas

plt.figure(figsize=(8,2))
ms_track.sort_values(ascending=False)[0:10].plot(kind='bar', color = 'green')

plt.title("Top 10 most listened to Tracks")
plt.xticks(rotation=55)
plt.xlabel("trackName")
plt.ylabel("Milliseconds heard")
plt.show()
```
